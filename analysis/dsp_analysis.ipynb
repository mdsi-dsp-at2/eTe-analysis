```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(color_codes=True)
```

```python
suicide_rate_data = pd.read_csv(r'/Users/priya/Downloads/crude_suicide_rates-2.csv')
#suicide_rate_data.head()
#suicide_rate_data.tail()
#suicide_rate_data.dtypes
#suicide_rate_data.count()
#print(suicide_rate_data.isnull().sum())
```

```python
suicide_rate_data['total']= suicide_rate_data.iloc[:, 3:11].sum(axis=1)
sort_total = suicide_rate_data.sort_values('total')
sort_total.head()
```

```python
Sex = sort_total["Country"].head()
Total = sort_total["total"].head()
# Figure Size 
fig = plt.figure(figsize =(10, 7)) 
  
# Horizontal Bar Plot 
plt.bar(Sex, Total) 
  
# Show Plot 
plt.show() 
```

```python
suicide_rate_data_female = suicide_rate_data[suicide_rate_data['Sex'] == "Female"]
sort_female = suicide_rate_data_female.sort_values('total')
sort_female.head()
```

```python
Country = sort_female["Country"].head(10) 
Total = sort_female["total"].head(10)
# Figure Size 
fig = plt.figure(figsize =(10, 7)) 
  
# Horizontal Bar Plot 
plt.barh(Country, Total) 
  
# Show Plot 
plt.show() 
```

```python
suicide_rate_data_male = suicide_rate_data[suicide_rate_data['Sex'] == "Male"]
sort_male = suicide_rate_data_male.sort_values('total')
sort_male.head()
```

```python
Country = sort_male["Country"].head(10) 
Total = sort_male["total"].head(10)
# Figure Size 
fig = plt.figure(figsize =(10, 7)) 
  
# Horizontal Bar Plot 
plt.barh(Country, Total) 
  
# Show Plot 
plt.show() 
```
