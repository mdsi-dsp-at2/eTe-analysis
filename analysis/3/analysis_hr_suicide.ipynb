```python
# Import the csv dataset

#hr = pd.read_csv('../../raw_data/human_resources.csv')
reader = SuicideDataReader()
human_resources = reader.read_data(SuicideRawData.HUMAN_RESOURCES)
```

```python
# changeing Country column type object to category

human_resources['country'] = human_resources['country'].astype('category')
```

```python
# changeing Year column type object to category

human_resources['year'] = human_resources['year'].astype('category')
```

```python
# statistics on Human resources for the dataset

human_resources.describe()
```

```python
# Plot the Psychiatrists versus Nurses service provisions in 2016 for all countries

sns.lmplot(data=human_resources, x='psychiatrists', y='nurses',height=10,scatter_kws={"s":200})
plt.show()
```
